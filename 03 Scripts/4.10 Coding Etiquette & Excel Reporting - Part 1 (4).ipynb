```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
path = r'C:\Users\Josepa Cusidó\092022 Instacart Basket Analysis'
```

```python
# Chunksize for a new dataset without the "Non-frequent customer"
ords_prods_customers = pd.DataFrame()
np.random.seed(4)
for chunk in pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'Orders_Products_Customers.csv'), chunksize=100000, header=0):
    dev = np.random.rand(len(chunk)) <= 0.5
    subset = chunk[dev]
    df_chunk = subset.loc[subset['frecuency_flag']!='Non-frequent customer'].dropna()
    ords_prods_customers = pd.concat([ords_prods_customers, df_chunk.copy()])
```
