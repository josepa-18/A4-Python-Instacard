```python
ords_prods_customers = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'Orders_Products_Customers.pkl'))
```

```python
# Create a Scatterplot between Age and spending power
sns.scatterplot(x = 'Age', y = 'income',data = ords_prods_customers)
```

The scatterplot above shows the relationship between age (0 to 80 years) and spending power (income).
- Most of the customers between 20 and 40 years old have an income of up to 200,000 USD. A few up to 400,000 USD.
- Most of the Customers aged 40 and above have an income of up to 300,000 USD. Among them, still many customers fall into the categoryof up to 200,000 USD. In this range, there are a few customers with an income of up to 500,000 USD.
